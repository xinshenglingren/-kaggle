```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from pandas import DataFrame
from sklearn.externals import joblib
```

```python
train_data = pd.read_csv('train.csv',names=range(14))
for i in range(train_data.shape[1]):
    if train_data[i].value_counts().keys()[0]=='?':
        x = train_data[i].value_counts().keys()[1]
    else:
        x = train_data[i].value_counts().keys()[0]
    train_data[i] = train_data[i].replace('?', x)
train_data
```

```python
test_data = pd.read_csv('test.csv',names=range(13))
for i in range(test_data.shape[1]):
    if test_data[i].value_counts().keys()[0]=='?':
        x = test_data[i].value_counts().keys()[1]
    else:
        x = test_data[i].value_counts().keys()[0]
    test_data[i] = test_data[i].replace('?', x)
test_data
```

```python
#取出文件中所需要的数据
train = train_data[[0,1,2,3,4,5]]
test = test_data[[0,1,2,3,4,5]]
#取出所有的性别
se= train[0]
#取出所有的婚姻状况
hunyin = train[1]
#取出所有的年龄
ege = train[2]
#取出所有的教育程度
edu = train[3]
#取出所有的职业
zhi = train[4]
#取出所有的工作年限
nianxian = train[5]
#取出所有预测数据中所有的性别
se_test = test[0]
#取出预测数据中所有的婚姻状况
hunyin_test = test[1]
#取出预测数据中所有的年龄
ege_test = test[2]
#取出预测数据中所有的教育程度
edu_test = test[3]
#取出所有预测数据中所有的职业
zhi_test = test[4]
#取出所有预测数据中所有的工作年限
nianxian_test = test_data[5]
```
